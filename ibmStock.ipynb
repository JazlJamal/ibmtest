```python
from bs4 import BeautifulSoup
import pandas as pd

# HTML content
html_content = """
<div id="style-1" style="background-color:#fff; height:510px; overflow:auto; margin: 30px 0px 30px 0px; padding:0px 30px 20px 0px; border:1px solid #dfdfdf;">
   <div class="col-xs-6">
      <table class="historical_data_table table">
         <thead>
            <tr>
               <th colspan="2" style="text-align:center">GameStop Annual Revenue<br><span style="font-size:14px;">(Millions of US $)</span></th>
            </tr>
         </thead>
         <tbody>
            <tr>
               <td style="text-align:center">2024</td>
               <td style="text-align:center">$5,273</td>
            </tr>
            <tr>
               <td style="text-align:center">2023</td>
               <td style="text-align:center">$5,927</td>
            </tr>
            <!-- more rows -->
         </tbody>
      </table>
   </div>
   <div class="col-xs-6">
      <table class="historical_data_table table">
         <thead>
            <tr>
               <th colspan="2" style="text-align:center">GameStop Quarterly Revenue<br><span style="font-size:14px;">(Millions of US $)</span></th>
            </tr>
         </thead>
         <tbody>
            <tr>
               <td style="text-align:center">2024-01-31</td>
               <td style="text-align:center">$1,794</td>
            </tr>
            <!-- more rows -->
         </tbody>
      </table>
   </div>
</div>
"""


soup = BeautifulSoup(html_content, 'html.parser')


annual_table = soup.find_all('table')[0]

data
quarterly_table = soup.find_all('table')[1]


def extract_data(table):
    rows = table.find_all('tr')
    data = []
    for row in rows[1:]:
        cols = row.find_all('td')
        year = cols[0].text.strip()
        revenue = cols[1].text.strip().replace('$', '').replace(',', '')
        data.append({'Year': year, 'Revenue (Millions of US $)': revenue})
    return data


annual_data = extract_data(annual_table)


quarterly_data = extract_data(quarterly_table)


gme_annual_revenue = pd.DataFrame(annual_data)
gme_quarterly_revenue = pd.DataFrame(quarterly_data)


print("Last five rows of GameStop Annual Revenue:")
print(gme_annual_revenue.tail())


print("\nLast five rows of GameStop Quarterly Revenue:")
print(gme_quarterly_revenue.tail())

```

```python
import yfinance as yf

def make_graph(data, title):
    import matplotlib.pyplot as plt
    
    plt.figure(figsize=(10, 6))
    plt.plot(data.index, data['Close'], color='blue', label='Close Price')
    plt.title(title)
    plt.xlabel('Date')
    plt.ylabel('Price (USD)')
    plt.legend()
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()


tesla_data = yf.download('TSLA', start='2021-01-01', end='2022-01-01')


tesla_data.reset_index(inplace=True)


make_graph(tesla_data, title='Tesla Stock Prices (2021)')

```
